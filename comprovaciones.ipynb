```python

# Definir las rutas de los archivos CSV para los años 2017 a 2012
rutas_archivos = {
    2017: "Datasetoriginales/sueldos-funcionarios-2017.csv",
    2016: "Datasetoriginales/sueldos-funcionarios-2016.csv",
    2015: "Datasetoriginales/sueldos-funcionarios-2015.csv",
    2014: "Datasetoriginales/SUELDOS_FUNCIONARIOS_2014.csv",
    2013: "Datasetoriginales/sueldos-funcionarios-2013.csv",
    2012: "Datasetoriginales/sueldos-funcionarios-2012.csv"
}

# Diccionario para almacenar los DataFrames de cada año
dataframes = {}

# Iterar sobre las rutas de los archivos y cargar los DataFrames
for year, ruta in rutas_archivos.items():
    try:
        df = pd.read_csv(ruta, delimiter=';')
        dataframes[year] = df
        print(f"Archivo para el año {year} cargado exitosamente.")
        print(df.head())
    except Exception as e:
        print(f"Error al cargar el archivo para el año {year}: {str(e)}")

# Definir los DataFrames df_2012 al df_2017 utilizando el diccionario dataframes
df_2012 = dataframes.get(2012)
df_2013 = dataframes.get(2013)
df_2014 = dataframes.get(2014)
df_2015 = dataframes.get(2015)
df_2016 = dataframes.get(2016)
df_2017 = dataframes.get(2017)

```

```python
# Definir los nombres de los DataFrames que se quieren verificar
nombres_dataframes = ["df_2024", "df_2023", "df_2022", "df_2020"]

# Crear un diccionario para almacenar las columnas de cada DataFrame
columnas_por_dataframe = {}

# Obtener las columnas de cada DataFrame
for nombre_df in nombres_dataframes:
    if nombre_df in locals():
        df = vars()[nombre_df]
        columnas_por_dataframe[nombre_df] = set(df.columns)
    else:
        print(f"El DataFrame {nombre_df} no está definido.")

# Verificar la compatibilidad entre los DataFrames
for i in range(len(nombres_dataframes)):
    for j in range(i + 1, len(nombres_dataframes)):
        df_name1 = nombres_dataframes[i]
        df_name2 = nombres_dataframes[j]
        if df_name1 in columnas_por_dataframe and df_name2 in columnas_por_dataframe:
            columns1 = columnas_por_dataframe[df_name1]
            columns2 = columnas_por_dataframe[df_name2]
            if columns1 == columns2:
                print(f"Los DataFrames {df_name1} y {df_name2} tienen columnas compatibles.")
            else:
                print(f"Los DataFrames {df_name1} y {df_name2} no tienen columnas compatibles.")
        else:
            print(f"No se pudieron verificar las columnas de {df_name1} y {df_name2}.")
```

```python
# Definir los nombres de los DataFrames que se quieren comparar
nombres_dataframes_2012_2017 = ["df_2012", "df_2013", "df_2014", "df_2015", "df_2016", "df_2017"]

# Crear un diccionario para almacenar las columnas de cada DataFrame
columnas_por_dataframe_2012_2017 = {}

# Obtener las columnas de cada DataFrame
for nombre_df in nombres_dataframes_2012_2017:
    if nombre_df in locals() and vars()[nombre_df] is not None:
        df = vars()[nombre_df]
        columnas_por_dataframe_2012_2017[nombre_df] = set(df.columns)
    else:
        print(f"El DataFrame {nombre_df} no está definido o es None.")

# Verificar la compatibilidad entre los DataFrames
for i in range(len(nombres_dataframes_2012_2017)):
    for j in range(i + 1, len(nombres_dataframes_2012_2017)):
        df_name1 = nombres_dataframes_2012_2017[i]
        df_name2 = nombres_dataframes_2012_2017[j]
        if df_name1 in columnas_por_dataframe_2012_2017 and df_name2 in columnas_por_dataframe_2012_2017:
            columns1 = columnas_por_dataframe_2012_2017[df_name1]
            columns2 = columnas_por_dataframe_2012_2017[df_name2]
            if columns1 == columns2:
                print(f"Los DataFrames {df_name1} y {df_name2} tienen columnas compatibles.")
            else:
                print(f"Los DataFrames {df_name1} y {df_name2} no tienen columnas compatibles.")
        else:
            print(f"No se pudieron verificar las columnas de {df_name1} y {df_name2}.")

```
